## 3.  왜 '>n' 그냥 앞 구간에 포함시켰는지에 대한 논리적 설명 

```python
import numpy as np

# 1. 대상 칼럼 리스트
range_columns = [
    '채취된 신선 난자 수',
    '수정 시도된 난자 수',
    '해동된 배아 수',
    '저장된 배아 수',
    '총 생성 배아 수'
]

# 2. 변환 함수 정의
def extract_max(val):
    if pd.isna(val):
        return -1  # 결측치도 -1 처리
    if isinstance(val, str):
        if '-' in val:
            try:
                return int(val.split('-')[1])     # 예: '1-5' → 5
            except:
                return -1
        elif val.startswith('>'):
            try:
                return int(val[1:]) +5              # 예: '>40' → 40
            except:
                return -1
        elif val.isdigit():
            return int(val)                       # 예: '0' → 0
    return -1  # 기타 예외 처리


# 3. 각 컬럼에 적용
for col in range_columns:
    df[col] = df[col].apply(extract_max)
```

```python
for col in range_columns:
    print(f"📌 {col}")
    print(df[col].value_counts().sort_index())
    print("-" * 60)
```

```python
avg_success = df.groupby('정자 기증자 나이')['임신 성공 확률'].mean()
# 2. 평균 확률이 가장 높은 나이 찾기
best_age = avg_success.idxmax()

best_age
```

```python

# 3. 절댓값 차이로 파생변수 생성
df['정자_나이차'] = (df['정자 기증자 나이'] - best_age).abs()
```

```python
# 정자_나이차 생성: -1이면 -1, 아니면 |정자 나이 - 38|
df['정자_나이차'] = df['정자 기증자 나이'].apply(lambda x: -1 if x == -1 else abs(x - 38))

# 난자_나이차 생성: -1이면 -1, 아니면 |난자 나이 - 28|
df['난자_나이차'] = df['난자 기증자 나이'].apply(lambda x: -1 if x == -1 else abs(x - 28))
```

```python
# 정자 기증자 나이별 임신 성공 확률 평균 계산
avg = df.groupby('정자 기증자 나이')['임신 성공 확률'].mean()

# 성공 확률이 가장 높은 나이 찾기 (단, 나이가 -1이 아닌 경우에 한해)
max_age = avg[avg.index != -1].idxmax()

# 파생변수 생성
def calc_age_distance(age):
    if age == -1:
        return -1
    else:
        return abs(age - max_age)

df['정자 나이차'] = df['정자 기증자 나이'].apply(calc_age_distance)
```
