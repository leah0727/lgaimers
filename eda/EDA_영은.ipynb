## 3.  ì™œ '>n' ê·¸ëƒ¥ ì• êµ¬ê°„ì— í¬í•¨ì‹œì¼°ëŠ”ì§€ì— ëŒ€í•œ ë…¼ë¦¬ì  ì„¤ëª… 

```python
import numpy as np

# 1. ëŒ€ìƒ ì¹¼ëŸ¼ ë¦¬ìŠ¤íŠ¸
range_columns = [
    'ì±„ì·¨ëœ ì‹ ì„  ë‚œì ìˆ˜',
    'ìˆ˜ì • ì‹œë„ëœ ë‚œì ìˆ˜',
    'í•´ë™ëœ ë°°ì•„ ìˆ˜',
    'ì €ì¥ëœ ë°°ì•„ ìˆ˜',
    'ì´ ìƒì„± ë°°ì•„ ìˆ˜'
]

# 2. ë³€í™˜ í•¨ìˆ˜ ì •ì˜
def extract_max(val):
    if pd.isna(val):
        return -1  # ê²°ì¸¡ì¹˜ë„ -1 ì²˜ë¦¬
    if isinstance(val, str):
        if '-' in val:
            try:
                return int(val.split('-')[1])     # ì˜ˆ: '1-5' â†’ 5
            except:
                return -1
        elif val.startswith('>'):
            try:
                return int(val[1:]) +5              # ì˜ˆ: '>40' â†’ 40
            except:
                return -1
        elif val.isdigit():
            return int(val)                       # ì˜ˆ: '0' â†’ 0
    return -1  # ê¸°íƒ€ ì˜ˆì™¸ ì²˜ë¦¬


# 3. ê° ì»¬ëŸ¼ì— ì ìš©
for col in range_columns:
    df[col] = df[col].apply(extract_max)
```

```python
avg_success = df.groupby('ì •ì ê¸°ì¦ì ë‚˜ì´')['ì„ì‹  ì„±ê³µ í™•ë¥ '].mean()
# 2. í‰ê·  í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ë‚˜ì´ ì°¾ê¸°
best_age = avg_success.idxmax()

best_age
```

```python

# 3. ì ˆëŒ“ê°’ ì°¨ì´ë¡œ íŒŒìƒë³€ìˆ˜ ìƒì„±
df['ì •ì_ë‚˜ì´ì°¨'] = (df['ì •ì ê¸°ì¦ì ë‚˜ì´'] - best_age).abs()
```

```python
# ì •ì_ë‚˜ì´ì°¨ ìƒì„±: -1ì´ë©´ -1, ì•„ë‹ˆë©´ |ì •ì ë‚˜ì´ - 38|
df['ì •ì_ë‚˜ì´ì°¨'] = df['ì •ì ê¸°ì¦ì ë‚˜ì´'].apply(lambda x: -1 if x == -1 else abs(x - 38))

# ë‚œì_ë‚˜ì´ì°¨ ìƒì„±: -1ì´ë©´ -1, ì•„ë‹ˆë©´ |ë‚œì ë‚˜ì´ - 28|
df['ë‚œì_ë‚˜ì´ì°¨'] = df['ë‚œì ê¸°ì¦ì ë‚˜ì´'].apply(lambda x: -1 if x == -1 else abs(x - 28))
```

```python
# ì •ì ê¸°ì¦ì ë‚˜ì´ë³„ ì„ì‹  ì„±ê³µ í™•ë¥  í‰ê·  ê³„ì‚°
avg = df.groupby('ì •ì ê¸°ì¦ì ë‚˜ì´')['ì„ì‹  ì„±ê³µ í™•ë¥ '].mean()

# ì„±ê³µ í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ë‚˜ì´ ì°¾ê¸° (ë‹¨, ë‚˜ì´ê°€ -1ì´ ì•„ë‹Œ ê²½ìš°ì— í•œí•´)
max_age = avg[avg.index != -1].idxmax()

# íŒŒìƒë³€ìˆ˜ ìƒì„±
def calc_age_distance(age):
    if age == -1:
        return -1
    else:
        return abs(age - max_age)

df['ì •ì ë‚˜ì´ì°¨'] = df['ì •ì ê¸°ì¦ì ë‚˜ì´'].apply(calc_age_distance)
```
